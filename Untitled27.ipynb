```python
import numpy as np
tokens = ["What", "are", "the", "symptoms", "of", "diabetes"]
```

```python
embeddings = np.array([
    [0.2, 0.1, 0.4, 0.7],
    [0.5, 0.3, 0.2, 0.1],
    [0.1, 0.6, 0.3, 0.2],
    [0.9, 0.4, 0.8, 0.5],
    [0.3, 0.2, 0.5, 0.6],
    [0.8, 0.7, 0.6, 0.9]
])
```

```python
W_Q = np.array([
    [0.1, 0.2, 0.3, 0.4],
    [0.2, 0.1, 0.4, 0.3],
    [0.3, 0.4, 0.1, 0.2],
    [0.4, 0.3, 0.2, 0.1]
])

W_K = np.array([
    [0.3, 0.1, 0.2, 0.4],
    [0.1, 0.4, 0.3, 0.2],
    [0.2, 0.3, 0.4, 0.1],
    [0.4, 0.2, 0.1, 0.3]
])

W_V = np.array([
    [0.2, 0.3, 0.1, 0.4],
    [0.3, 0.2, 0.4, 0.1],
    [0.4, 0.1, 0.3, 0.2],
    [0.1, 0.4, 0.2, 0.3]
])
```

```python
Q = np.dot(embeddings, W_Q)
K = np.dot(embeddings, W_K)
V = np.dot(embeddings, W_V)
```

```python
d_k = K.shape[1]
scores = np.dot(Q, K.T) / np.sqrt(d_k)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
```

```python
attention_weights = softmax(scores)
```
